```python
#converting annual trips to a list:
capacity1 = pd.read_excel(open('P3_DCLocation.xlsx','rb'), sheet_name='DCdata') #Change to your file name and sheet name
capacity1
```

```python
capacity = capacity1["capacity"].tolist()
capacity
```

```python
fixed_costs = capacity1['fixedcost']
fixed_costs
```

```python
cities = capacity1['location'].tolist()
cities
```

```python
# Declare a concrete model
model = ConcreteModel()
num_dc_loc = 76 #possible SC Location, (columns) index by i
num_city_loc = 23 #possible ciy location, (rows) index them by j
#Declare the decision variables
model.x = Var(range(num_dc_loc), domain = Binary) #Locate SC in that city? Indexed by i
model.y = Var(range(num_city_loc), range(num_dc_loc), domain = NonNegativeReals) # assign a city to a specific SC? Indexed by j,i

#Constraint #1: The amount produced needs to be greater than or equal to demand for each region...
model.Demand = ConstraintList()
for j in range(num_dc_loc):
    model.Demand.add(expr = sum(model.y[i,j] for i in range(num_city_loc)) >= demand[j])

#Constraint #2: The amount produced cannot exceed capacity....
model.Capacity = ConstraintList()
for i in range(num_city_loc):
    model.Capacity.add(expr = sum(model.y[i,j] for j in range(num_dc_loc)) <= model.x[i]*capacity[i])
    
#Set the model Objective...
variable_cost = sum(model.y[i,j]*distances[i][j] for i in range(num_city_loc) for j in range(num_dc_loc))
fixed_cost = sum(model.x[i]*fixed_costs[i] for i in range(num_city_loc))
total_costs = fixed_cost + variable_cost
model.Objective = Objective(expr = total_costs, sense = minimize)
```

```python
opt = SolverFactory('glpk')
opt.options['mipgap'] = .001
results = opt.solve(model, tee=True)
```
